```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from pathlib import Path
```

```python
April_may_df = pd.read_csv('Classes April-May 2018.csv')
April_may_df.head()
```

```python
June_df =  pd.read_csv('Classes June 2018.csv')
June_df.head()
```

```python
# checking null values
April_may_df.isnull().sum() / len(April_may_df)* 100
```

```python
# checking null values for june data
June_df.isnull().sum() / len(April_may_df)* 100
```

```python
# dropping null values as it has no price mentioned in it
df1 = April_may_df.dropna()
df2 = June_df.dropna()
```

```python
df1.drop_duplicates()
df2.drop_duplicates()
```

```python
df1 = df1.rename(columns={'BookingEndDateTime (Month / Day / Year)': 'BookingEndDateTime', 'Price (INR)' : 'PriceINR'})
df2 = df2.rename(columns={'BookingEndDateTime (Month / Day / Year)': 'BookingEndDateTime', 'Price (INR)' : 'PriceINR'})
```

```python
df1['BookingEndDateTime'] = pd.to_datetime(df1['BookingEndDateTime'])
df2['BookingEndDateTime'] = pd.to_datetime(df2['BookingEndDateTime'])
```

```python
df1.info()
```

```python
df2.info()
```

```python
# Correct data types
df1['MaxBookees'] = df1['MaxBookees'].astype(int)
df1['Number Booked'] = df1['Number Booked'].astype(int)
df1['PriceINR'] = pd.to_numeric(df1['PriceINR'].astype(int), errors='coerce')
df2['MaxBookees'] = df2['MaxBookees'].astype(int)
df2['Number Booked'] = df2['Number Booked'].astype(int)
df2['PriceINR'] = pd.to_numeric(df2['PriceINR'], errors='coerce')
```

```python
# Merge datasets using a common key (e.g., Class ID, Studio ID).
df_merged = pd.concat([df1, df2], ignore_index=True)
df_merged
```

```python
df_merged['ActivityDescription'] = df_merged['ActivityDescription'].astype(str).str.strip().str.lower()
df_merged
```

```python
df_merged.isnull().sum()
```

```python
# Extract hour from BookingStartTime
df_merged['Hour'] = pd.to_numeric(df_merged['BookingStartTime'].str.extract(r'^(\d{1,2}):', expand=False), errors='coerce')
```

```python
# Extract date components for time series analysis
df_merged['Month'] = df_merged['BookingEndDateTime'].dt.month
df_merged['Week'] = df_merged['BookingEndDateTime'].dt.isocalendar().week
df_merged['DayOfWeek'] = df_merged['BookingEndDateTime'].dt.day_name()
df_merged['DayOfWeekNum'] = df_merged['BookingEndDateTime'].dt.dayofweek
```

```python
# Occupancy rate
df_merged['occupancy_rate'] = df_merged['Number Booked'] / df_merged['MaxBookees']
df_merged['occupancy_rate'] = df_merged['occupancy_rate']
df_merged
```

```python
# QA checks
miss = df_merged.isna().mean().sort_values(ascending=False).reset_index()
miss.columns = ['column','fraction_missing']
dup_count = df_merged.duplicated().sum()
mixed_types = {c: df_merged[c].dropna().apply(lambda x: type(x)).value_counts().to_dict() for c in df_merged.columns if df_merged[c].dropna().apply(lambda x: type(x)).nunique() > 1}
num_stats = {c: {"count": int(df_merged[c].count()), "min": float(df_merged[c].min()), "max": float(df_merged[c].max()), "mean": float(df_merged[c].mean()), "std": float(df_merged[c].std())} for c in df_merged.select_dtypes(include=[np.number]).columns}
print("="*80)
print("Checking for any missing values in column =", miss)
print("="*80)
print("Checking for duplicate rows =", dup_count)
print("="*80)
num_stats_df = pd.DataFrame(num_stats).T.reset_index()
num_stats_df
```

```python
# Normalize numerical features
scaler = StandardScaler()
num_cols = ['Number Booked', 'MaxBookees', 'PriceINR']
df_merged[num_cols] = scaler.fit_transform(df_merged[num_cols]).round(2)
df_merged
```

```python
# cleaning actions
cleaned = df_merged.copy()
all_null_cols = [c for c in cleaned.columns if cleaned[c].isna().all()]
cleaned = cleaned.drop(columns=all_null_cols)
cleaned = cleaned.drop_duplicates().reset_index(drop=True)
```

```python
DATA_DIR = Path('.') # Define DATA_DIR as the current directory

# save
cleaned.to_excel(DATA_DIR / "cleaned_dataset.xlsx", index=False)
cleaned.to_csv(DATA_DIR / "cleaned_dataset.csv", index=False)
```
